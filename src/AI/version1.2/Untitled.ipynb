```python
def ema(series, periods, fillna=False):
    if fillna:
        return series.ewm(span=periods, min_periods=0).mean()
    return series.ewm(span=periods, min_periods=periods).mean()


def rsi(close, n=14, forecast_len=20, fillna=False):
    """Relative Strength Index (RSI)
    Compares the magnitude of recent gains and losses over a specified time
    period to measure speed and change of price movements of a security. It is
    primarily used to attempt to identify overbought or oversold conditions in
    the trading of an asset.
    https://www.investopedia.com/terms/r/rsi.asp
    Args:
        close(pandas.Series): dataset 'Close' column.
        n(int): n period.
        fillna(bool): if True, fill nan values.
    Returns:
        pandas.Series: New feature generated.
    """
    diff = close.diff(forecast_len)
    which_dn = diff < 0

    up, dn = diff, diff*0
    up[which_dn], dn[which_dn] = 0, -up[which_dn]

    emaup = ema(up, n, fillna)
    emadn = ema(dn, n, fillna)

    rsi = 100 * emaup / (emaup + emadn)
    if fillna:
        rsi = rsi.replace([np.inf, -np.inf], np.nan).fillna(50)
    return pd.Series(rsi, name='rsi')
```

```python
def rsi2(close, n=14, forecast_len=20, fillna=False):
    
    delta = close.diff(forecast_len)
# Get rid of the first row, which is NaN since it did not have a previous 
# row to calculate the differences
    delta = delta[forecast_len:] 

# Make the positive gains (up) and negative gains (down) Series
    up, down = delta.copy(), delta.copy()
    up[up < 0] = 0
    down[down > 0] = 0

# Calculate the EWMA
    roll_up1 = up.ewm(n).mean()
    roll_down1 = down.abs().ewm(n).mean()

# Calculate the RSI based on EWMA
    RS1 = roll_up1 / roll_down1
    RSI1 = 100.0 - (100.0 / (1.0 + RS1))

# Calculate the SMA
    roll_up2 = up.rolling( n).mean()
    roll_down2 = down.abs().rolling(n).mean()

# Calculate the RSI based on SMA
    RS2 = roll_up2 / roll_down2
    RSI2 = 100.0 - (100.0 / (1.0 + RS2))
    
    return RSI1, RSI2
```

```python
def atr(df, open_col, high_col, low_col, close_col, forecast_len=20, n=14):
    df['max'] = df[['BNB.o', 'BNB.h', 'BNB.l', 'BNB.p']].max(axis=1)
    df['min'] = df[['BNB.o', 'BNB.h', 'BNB.l', 'BNB.p']].min(axis=1)
    df['val1'] = df['max'] - df['min']
    df['val2'] = abs(df['max'] - df['BNB.p'].shift(forecast_len))
    df['val3'] = abs(df['min'] - df['BNB.p'].shift(forecast_len))
    df['tr'] = df[['val1', 'val2', 'val3']].max(axis=1)
    df['atr'] = None
    for i in range(forecast_len):
        df['atr'].iloc[i::forecast_len] = df['tr'].iloc[i::forecast_len].ewm(ignore_na=False, min_periods=0, com=n, adjust=True).mean()
    
    df = df.drop(['max', 'min', 'val1', 'val2', 'val3', 'tr'], axis=1)
        
    return df
```
