```python
import numpy as np
import pandas as pd
import os
```

```python
# Constants
seq_len = 1000
rand_init = 5000
forecast_len = 20
```

```python
dfo = pd.read_csv(os.path.join('21_coins_tot.csv'), index_col = 0)
```

```python
df = dfo.iloc[:, :-seq_len-forecast_len]
```

```python
ref = dfo.iloc[:, -seq_len-forecast_len:-forecast_len]
```

```python
#rp = ref.iloc[:, rand_init:rand_init+seq_len]
```

```python
rows = seq_len*len(df.index)
cols = int(df.shape[1]/seq_len)
df = df.iloc[:,:seq_len*cols].values.reshape((rows, cols), order = 'F')
```

```python
rpf = ref.values.flatten('F')
```

```python
rpf.shape
```

```python
correls = [np.corrcoef(df[:,a], rpf)[1,0] for a in range(df.shape[1])]
```

```python
correls
```

```python
np.argmax(correls)
```

```python
vec1 = dfo.iloc[:, np.argmax(correls)*seq_len+forecast_len]
```

```python
vec2 = dfo.iloc[:, -1]
```

```python
np.corrcoef(vec1, vec2)
```
